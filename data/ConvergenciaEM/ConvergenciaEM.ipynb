```python
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
def ConvEM(doc,sigma,alpha,m,sigmaInt,alphaInt,mInt):
    df = pd.read_csv(doc, sep=',',header=None)
    indices_a_eliminar = list(range(0, len(df), 12))
    df = df.drop(indices_a_eliminar)
    dfs = np.array_split(df, 1000)
    dfs = [dfs[i].rename(columns=dfs[i].iloc[0]).drop(dfs[i].index[0])for i in range(len(dfs))] 

    sigmaEMdata = [np.mean([float(dfs[i][' sigma_hat'].to_numpy()[k]) for i in range(1000)]) for k in range(10)]
    alphaEMdata = [np.mean([float(dfs[i]['alpha_hat'].to_numpy()[k]) for i in range(1000)]) for k in range(10)]
    mEMdata = [np.mean([float(dfs[i]['m_hat'].to_numpy()[k]) for i in range(1000)]) for k in range(10)]

    fig, axes = plt.subplots(nrows=3,figsize=(10, 8))

    axes[0].scatter(range(1,11), sigmaEMdata, color='darkblue',s=50,label= '$\hat{\sigma}_{EM}$',marker="8")
    axes[0].axhline(y=sigma, color='red', linestyle='--')
    axes[0].tick_params(axis='both', which='major', labelsize=20)
    axes[0].set(xticklabels=[])
    axes[1].set_ylim(sigmaInt) ################ Cambio de escala ############
    axes[0].set_ylabel('$\hat{\sigma}_{EM}$',fontsize=30)

    axes[1].scatter(range(1,11), alphaEMdata, color='darkgreen',s=50,label= '$\hat{\alpha}_{EM}$',marker="8")
    axes[1].axhline(y=alpha, color='red', linestyle='--')
    axes[1].tick_params(axis='both', which='major', labelsize=20)
    axes[1].set(xticklabels=[])
    axes[1].set_ylim(alphaInt) ################ Cambio de escala ############
    axes[1].set_ylabel('$\hat{Î±}_{EM}$',fontsize=30)

    axes[2].scatter(range(1,11), mEMdata, color='darkred',s=50,label= '$\hat{m}_{EM}$',marker="8")
    axes[2].axhline(y=m, color='red', linestyle='--')
    axes[2].set_ylabel('$\hat{m}_{EM}$',fontsize=30)
    axes[2].tick_params(axis='both', which='major', labelsize=20)
    axes[1].set_ylim(mInt) ################ Cambio de escala ############
    axes[2].set_xlabel('Iterations',fontsize=30) 
```
